```python
import pandas as pd
from pmdarima import auto_arima
from tqdm import tqdm 
import numpy as np

train_data = pd.read_csv(r"C:\Users\imaja\OneDrive\Desktop\MBA 2022 -2024\T6\Capstone\Datasets\transposed.csv")

# Assuming 'Year' is the column containing years as strings
train_data['Year'] = pd.to_datetime(train_data['Year'], format='%Y')
```

```python
train_data
```

```python
# Assuming 'Year' is the column containing years as strings
train_data['Year'] = pd.to_datetime(train_data['Year'], format='%Y')
train_data.set_index('Year', inplace=True)

# Get column names excluding 'Year'
value_columns = [col for col in train_data.columns if col != 'Year']

# Iterate through columns '0' to '2417' with tqdm
for value_column in tqdm(value_columns, desc="Processing Columns"):
    # Split the data into train and test sets (2010 to 2017 as train, exclude 2018 and 2019)
    train_set = train_data[value_column]['2010':'2017']
    test_set = train_data[value_column]['2018':'2019']

    # Perform AutoARIMA for the training set
    model = auto_arima(train_set, seasonal=True, stepwise=True)

    # Make future predictions for the test set (adjust the 'n_periods' parameter as needed)
    future_forecast = model.predict(n_periods=len(test_set))

    # Add new rows for 2018-01-01 and 2019-01-01 with forecasted results to the dataframe
    train_data.loc[pd.to_datetime('2018-01-01'), value_column] = future_forecast[0]
    train_data.loc[pd.to_datetime('2019-01-01'), value_column] = future_forecast[1]
```
